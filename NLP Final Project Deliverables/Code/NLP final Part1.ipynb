```python
paper=[]
for x in data['body_text']:
    text = x['text']
    paper.append(text)
    #print(text, '\n')
    print (paper)
```

```python
import re
```

```python
#This is for words contains special symbols
delimiter1_list=[]
for x in range(18):
    delimiter1= re.findall(r"[a-zA-Z]+[.(?*)-/:{}<>%=;,+!?*]+[a-zA-Z.()-/:{}<>%=;,+!?*]+",paper[x])
    delimiter1_list.extend(delimiter1)
delimiter1_list=list(set(delimiter1_list))
delimiter1_list
```

```python
#This is for words contains Roman Numeral
single1_list=[]
for x in range(18):
    single1= re.findall(r'\w*I\b|\w*V\b|\w*X\b|\w*D\b|\w*L\b|\w*CM\b',paper[x])
    single1.extend(single1)
single1_list=list(set(single1_list))
single1_list
```

```python
#This is for words contains number
single2_list=[]

for x in range(18):
    single2= re.findall(r"\w*[0-9]+",paper[x])
    single2_list.extend(single2)
single2_list=list(set(single2_list))
single2_list
```

```python
#This is for single letter
single3_list=[]
for x in range(18):
    single3= re.findall(r'(?i)\b[a-z]\b',paper[x])
    single3_list.extend(single3)
single3_list=list(set(single3_list))
single3_list
```

```python
#This is for greek letter
single4_list=[]
for x in range(18):
    single4= re.findall(r'\w*alpha\b|\w*beta\b|\w*gamma\b|\w*delta\b|\w*epsilon\b|\w*theta\b|\w*kappa\b|\w*lambda\b|\w*sigma\b|\w*mu\b',paper[x])
    single4_list.extend(single4)
single4_list=list(set(single4_list))
single4_list
```

```python
#This is for Abbreviation
abbre1_list=[]
for x in range(18):
    abbre1=re.findall(r'\b(?:[a-z]*[A-Z|0-9][a-z|0-9]*){2,}',paper[x])
    abbre1_list.extend(abbre1)
abbre1_list=list(set(abbre1_list))
abbre1_list
```

```python
#This is for BioRegular
bioregular_list=[]
for x in range(18):
    bioregular=re.findall(r'\w*ase\b|\w*gen\b|\w*in\b|\w*al\b|\w*um\b|\w*ide\b',paper[x])
    bioregular_list.extend(bioregular)
bioregular_list=list(set(bioregular_list))
bioregular_list
```

```python
#This is for lowercase
lowercase_list=[]
for x in range(18):
    lowercase=re.findall(r'\b[^A-Z\s\d]+\w{4,}\b',paper[x])
    lowercase_list.extend(lowercase)
lowercase_list=list(set(lowercase_list))
lowercase_list
```

```python
#This is for uppercase
uppercase_list=[]
for x in range(18):
    uppercase=re.findall(r'[A-Z][a-z]+\w{4,}\b',paper[x])
    uppercase_list.extend(uppercase)
uppercase_list=list(set(uppercase_list))
uppercase_list
```

```python
# This is for ATCG Sequence
sequences_list=[]
for x in range(18):
    sequences=re.findall(r'\w+[A]+[C]+[T]+[G]+',paper[x])
    sequences_list.extend(sequences)
sequences_list=list(set(sequences_list))
sequences_list
```

```python
# combine all the regex result
protein_words=delimiter1_list+single2_list+single3_list+single4_list+abbre1_list+bioregular_list+lowercase_list+uppercase_list+sequences_list
# remove duplicates
protein_words=list(set(protein_words))
```
